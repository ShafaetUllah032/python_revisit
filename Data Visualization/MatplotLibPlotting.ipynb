```python
#from a pandas dataframe

batsman=pd.read_csv('G:\WORK\python\Data Visualization\dataset\sharma-kohli.csv')
batsman.sample(5)
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
#labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Career comp Kohli and Sharma')
plt.xlabel('Session')
plt.ylabel('Runs Score')
```

```python
#colors(hex) and line(width and style) and marker(size)
#https://htmlcolorcodes.com/

plt.plot(batsman['index'],batsman['V Kohli'],color='#CF4AC1')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#30B609')

plt.title('Career comp Kohli and Sharma')
plt.xlabel('Session')
plt.ylabel('Runs Score')
```

```python
#using line style
plt.plot(batsman['index'],batsman['V Kohli'],color='#CF4AC1',linestyle='dashed',linewidth=3,marker='D',markersize='10')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#30B609',linestyle='dashdot',linewidth=5,marker='o',markersize='10')

plt.title('Career comp Kohli and Sharma')
plt.xlabel('Session')
plt.ylabel('Runs Score')
```

```python
#legend -> location

plt.plot(batsman['index'],batsman['V Kohli'],color='#CF4AC1',linestyle='dashed',linewidth=3,marker='D',markersize='10',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#30B609',linestyle='dashdot',linewidth=5,marker='o',markersize='10',label='Rohit')

plt.title('Career comp Kohli and Sharma')
plt.xlabel('Session')
plt.ylabel('Runs Score')
plt.legend(loc='best')
```

```python
# grid
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
```

```python
# show
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()

plt.show()
```

```python
# size
tips = sns.load_dataset('tips')


# slower
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)
```

```python
# scatterplot using plt.plot
# faster
plt.plot(tips['total_bill'],tips['tip'],'o')
```

### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='black')
```

```python
# horizontal bar chart
plt.barh(colors,children,color='black')
```

```python
# color and label
df = pd.read_csv('G:\WORK\python\Data Visualization\dataset\Batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
# dataset
df = pd.read_csv('G:\WORK\python\Data Visualization\dataset\gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
